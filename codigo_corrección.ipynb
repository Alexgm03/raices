```python
def secant_method(f, x0, x1, tol=1e-4, max_iter=100):
    x_prev = x0
    x_curr = x1

    f_prev = f(x_prev)
    f_curr = f(x_curr)

    iter_count = 0

    while abs(f_curr) > tol and iter_count < max_iter:

        if f_curr == f_prev:
            print("Error: división por cero")
            return None, iter_count

        x_next = x_curr - f_curr * (x_curr - x_prev) / (f_curr - f_prev)

        x_prev = x_curr
        f_prev = f_curr

        x_curr = x_next
        f_curr = f(x_curr)

        iter_count += 1

    return x_curr, iter_count

i = 0

def func(x):
    global i
    i += 1
    y = x**3 - 3*x**2 + x - 1
    print(f"llamada {i}\t x={x:.5f}\t y={y:.5f}")
    return y


raiz, iteraciones = secant_method(func, 2, 3)

print(raiz, iteraciones)
```
